```python
def irrigation_safety(row):
    cond_ok = row['Conductivity'] < 3000
    tds_ok = row['Solids'] < 3000
    ph_ok = (row['ph'] >= 6.5 and row['ph'] <= 8.4)

    if cond_ok and tds_ok and ph_ok:
        return 1   # Safe
    else:
        return 0   # Unsafe

df['Irrigation_Safety'] = df.apply(irrigation_safety, axis=1)
```

```python
#Handle missing values in 'ph' column
df['ph'].fillna(df['ph'].mean(), inplace=True) #It fills all missing (NaN) values in the ph column of the DataFrame df with the mean value of that same column.
#inplace=True->This updates the DataFrame df directly, without creating a new copy.
```

```python
#Handle missing values in 'Sulfate' column
df['Sulfate'].fillna(df['Sulfate'].median(), inplace=True)
```

```python
#Handle missing values in 'Trihalomethanes' column
df['Trihalomethanes'].fillna(df['Trihalomethanes'].median(), inplace=True)
```

```python
#Train–Test Split
```

```python
from sklearn.model_selection import train_test_split
# What is train_test_split?
# It’s a function from scikit-learn (sklearn) used to split your dataset into training and testing parts.
# The idea:
# Training set → used to train the machine learning model
# Testing set → used to evaluate how well the model performs on unseen data
```

```python
X = df.drop(axis=1, columns=['Irrigation_Safety'])#Removes the 'Irrigation_Safety' column from df to get only the input features.
y = df['Irrigation_Safety']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=42, stratify=y)
```

```python
# X_train → features used to train the model (70% of total rows)

# X_test → features used to test the model (30% of total rows)

# y_train → corresponding target values for X_train

# y_test → corresponding target values for X_test
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
#Standardization
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
# StandardScaler transforms your data so that each feature (column) has:
# Mean = 0
# Standard deviation = 1
# This process is called standardization or z-score normalization.
```

```python
X_train_scaled= scaler.fit_transform(X_train)
```

```python
# This single line does two things at once:
# fit(X_train) → Calculates the mean and standard deviation of each feature (column) in your training data.
# transform(X_train) → Uses those calculated statistics to standardize (scale) each value in X_train using the formula:
# 𝑋' = (X - mean)/std
# So the final result X_train_scaled contains data where each column has mean = 0 and std = 1.
```

```python
X_test_scaled=scaler.transform(X_test) #Scales test data using the same parameters from training
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
#Model Training and Evaluation
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier()
}
```

```python
results = []

for name,model in models.items():
    model.fit(X_train_scaled,y_train)

    y_pred_train = model.predict(X_train_scaled)
    y_pred_test = model.predict(X_test_scaled)

    results.append({
        "Model": name,
        "Train Accuracy": accuracy_score(y_train, y_pred_train),
        "Train Precision": precision_score(y_train, y_pred_train),
        "Train Recall": recall_score(y_train, y_pred_train),
        "Train F1": f1_score(y_train, y_pred_train),

        "Test Accuracy": accuracy_score(y_test, y_pred_test),
        "Test Precision": precision_score(y_test, y_pred_test),
        "Test Recall": recall_score(y_test, y_pred_test),
        "Test F1": f1_score(y_test, y_pred_test)
    })
```

```python
results_df = pd.DataFrame(results)
results_df = results_df.round(4)
```

```python
results_df = results_df.sort_values(by="Test Accuracy", ascending=False).reset_index(drop=True)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params_grid = {
    "Logistic Regression": {
        'penalty': ['l1','l2'],
        'C': [0.01, 0.1, 1, 10, 100],
        'solver': ['liblinear', 'saga'],
        'max_iter': [200, 500]
    },
    "Random Forest": {
        'n_estimators': [100, 200, 300, 500],
        'criterion': ['gini', 'entropy'],
        'max_depth': [None, 5, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'max_features': ['sqrt', 'log2'],
        'bootstrap': [True, False]
    }
}
```

```python
best_models = []

for name, model in models.items():
    print(f"Tuning {name}:")

    random_search = RandomizedSearchCV(
        estimator=model,
        param_distributions=params_grid[name],
        n_iter=20,
        scoring='accuracy',
        cv=3,
        n_jobs=1,
        verbose=1,
        random_state=42
    )

    random_search.fit(X_train_scaled, y_train)

    best_models.append({
        "Model": name,
        "Best Params": random_search.best_params_,
        "Best CV Score": random_search.best_score_,
        "Train Accuracy": accuracy_score(y_train, random_search.predict(X_train_scaled)),
        "Test Accuracy": accuracy_score(y_test, random_search.predict(X_test_scaled))
    })
```

```python
best_models
```

```python
results_df = pd.DataFrame(best_models)
#results_df = results_df.dropna(subset=["Best CV Score"])
results_df = results_df.round(4)
results_df = results_df.sort_values(by="Test Accuracy", ascending=False).reset_index(drop=True)
```
